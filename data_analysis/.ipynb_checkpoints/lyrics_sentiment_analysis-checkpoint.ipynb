# LSTM을 이용한 음악 가사 감성 분석

LSTM(Long Short-Term Memory)을 이용하여 음악 가사의 감정을 분류하는 머신러닝 모델입니다. 이 프로젝트의 목적은 음악 가사를 분석하여 해당 가사가 전달하는 감정을 긍정, 부정, 중립으로 분류하는 것입니다.

### 1. 필요한 라이브러리 임포트

여러 필요한 라이브러리들이 임포트합니다. 이들은 데이터 처리, 모델 구축, 시각화 등에 사용됩니다.
Okt는 KoNLPy(한국어 처리를 위한 파이썬 라이브러리)의 클래스입니다. 이는 텍스트를 토큰화하는데 사용됩니다.

```python
from __future__ import annotations
from keras.preprocessing.text import Tokenizer
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Embedding, LSTM, Dropout
from keras.callbacks import EarlyStopping
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from konlpy.tag import Okt
from keras.utils import pad_sequences
```

### 2. 데이터 수집 및 준비

훈련 및 테스트 데이터가 CSV 파일에서 로드됩니다. 본 코드에서는 다음과 같은 데이터셋을 사용하였습니다:
- 훈련 데이터 (train.csv): 음악 가사와 이에 대응하는 감정 라벨(긍정, 부정, 중립)이 포함되어 있습니다.
- 테스트 데이터 (test.csv): 모델 평가를 위한 별도의 가사와 감정 라벨을 포함합니다.
- 음악 정보 (music_info.csv): 각 가사에 해당하는 음악의 제목 정보를 포함합니다.

데이터의 라벨 분포를 막대 그래프로 시각화합니다. 이를 통해 데이터의 불균형을 파악하고, 모델 설계 시 이를 고려할 수 있습니다.

```python
# 토크나이저 생성
tokenizer = Okt()

# 데이터셋 불러오기
train_data = pd.read_csv("../data/train.csv")
test_data = pd.read_csv("../data/test.csv")
title_name=pd.read_csv("../data/music_info.csv")

# 훈련 데이터의 라벨 분포 시각화
train_data['label'].value_counts().plot(kind='bar')
plt.show()

# 테스트 데이터의 라벨 분포 시각화
test_data['label'].value_counts().plot(kind='bar')
plt.show()

# 훈련 및 테스트 데이터의 라벨 분포 출력
print(train_data.groupby('label').size().reset_index(name='count'))
print(test_data.groupby('label').size().reset_index(name='count'))
```

### 3. 데이터 전처리

- 불용어 제거: stopwords-ko.txt 파일을 참조하여 불필요한 단어를 제거함으로써 데이터의 질을 개선했습니다.
- 토큰화: 한국어 가사의 효과적인 분석을 위해 Okt 토크나이저를 사용하여 텍스트를 텍스트를 더 작은 단위, 즉 '토큰'으로 분해하였습니다.

```python
# 불용어 사전 로드
file_path="../words/stopwords-ko.txt"

with open(file_path,encoding='utf-8') as f:
  lines = f.readlines()
lines = [line.rstrip('\n') for line in lines]

stopwords = lines

# 훈련 데이터 전처리
okt = Okt()
X_train = []
for lyrics in train_data['lyric']:
  temp_X = []
  temp_X = okt.morphs(lyrics, stem=True) # 토큰화
  temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
  X_train.append(temp_X)

# 테스트 데이터 전처리
X_test = []
for lyrics in test_data['lyrics']:
  temp_X = []
  temp_X = okt.morphs(lyrics, stem=True) # 토큰화
  temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
  X_test.append(temp_X)

# 토큰화된 데이터의 일부 출력
print(X_train[100:103])
print(X_test[100:103])
```

```python
# 토큰화 설정 및 토큰화 진행
max_words = 35000
tokenizer = Tokenizer(num_words = max_words)
tokenizer.fit_on_texts(X_train)
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test) 
print(X_train[100:103])
print(X_test[100:103])
```

```python
# 훈련 데이터의 길이 분석
print("가사의 최대 길이 : ", max(len(l) for l in X_train))
print("가사의 평균 길이 : ", sum(map(len, X_train))/ len(X_train))
plt.hist([len(s) for s in X_train], bins=50)
plt.xlabel('length of Data')
plt.ylabel('number of Data')
plt.show()
```

```python
# 테스트 데이터의 길이 분석
print("가사의 최대 길이 : ", max(len(l) for l in X_test))
print("가사의 평균 길이 : ", sum(map(len, X_test))/ len(X_test))
plt.hist([len(s) for s in X_test], bins=50)
plt.xlabel('length of Data')
plt.ylabel('number of Data')
plt.show()
```

- 원-핫 인코딩: 훈련 데이터와 테스트 데이터의 라벨을 처리합니다.
    - 라벨 1 : 긍정 [0, 0, 1]
    - 라벨 0 : 중립 [0, 1, 0]
    - 라벨 -1 : 부정 [1, 0, 0]

```python
# 라벨 데이터 준비
y_train = []
y_test = []

for i in range(len(train_data['label'])):
  if train_data['label'].iloc[i] == 1:
    y_train.append([0, 0, 1])
  elif train_data['label'].iloc[i] == 0:
    y_train.append([0, 1, 0])
  elif train_data['label'].iloc[i] == -1:
    y_train.append([1, 0, 0])

for i in range(len(test_data['label'])):
  if test_data['label'].iloc[i] == 1:
    y_test.append([0, 0, 1])
  elif test_data['label'].iloc[i] == 0:
    y_test.append([0, 1, 0])
  elif test_data['label'].iloc[i] == -1:
    y_test.append([1, 0, 0])

y_train = np.array(y_train)
y_test = np.array(y_test)

print(y_train[100:103])
print(y_test[100:103])
```

### 4. 모델 구성

Sequential 모델을 사용하며, 여러 레이어를 포함합니다:
- Embedding 레이어: 입력 시퀀스를 고정 크기의 밀집 벡터로 변환합니다. 여기서 max_words는 35000으로 설정되어 있으며, 각 단어는 100차원의 벡터로 표현됩니다.
- LSTM 레이어: 128개의 유닛을 가진 LSTM 레이어를 사용합니다. LSTM은 장단기 기억을 모두 고려하는 순환 신경망(RNN)의 한 종류로, 시퀀스 데이터의 맥락을 잘 파악하는 데 효과적입니다.
- Dropout 레이어: 과적합을 방지하기 위해 뉴런의 50%를 무작위로 비활성화합니다.
- Dense 레이어: 최종 출력 레이어로, softmax 활성화 함수를 사용하여 각 클래스에 대한 확률을 출력합니다.

```python
# 데이터 길이 맞추기
max_len = 12 # 전체 데이터의 길이를 12로 맞춤
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = pad_sequences(X_test, maxlen=max_len)

# 모델 구성
model = Sequential()
model.add(Embedding(max_words, 100))
model.add(LSTM(64,recurrent_dropout=0.2))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))
```

### 5. 모델 학습 및 검증

- 컴파일: Adam 최적화 알고리즘과 범주형 크로스 엔트로피 손실 함수를 사용했습니다.
- 조기 중단: 과적합을 방지하기 위해 val_loss의 개선이 없는 경우 3회의 에포크 후 훈련을 중단했습니다.
- 학습: 모델은 총 20 에포크 동안 훈련되었으며, 검증 데이터셋을 사용하여 각 에포크의 성능을 모니터링했습니다.

```python
# 모델 컴파일
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 조기 종료 설정
early_stopping = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=5)

# 모델 학습
history = model.fit(X_train, y_train, epochs=20, validation_split=0.2, callbacks = [early_stopping])
```

### 6. 모델 테스트 및 예측

테스트 데이터로 모델의 정확도를 평가합니다. 모델은 테스트 데이터셋에서 약 83.19%의 정확도를 달성했습니다. 테스트 데이터의 가사 샘플과 실제 라벨을 비교한 결과, 모델은 다소 우수하지 못한 성능을 보였습니다. 이를 개선하기 위해 모델의 Fine Tuning을 진행할 예정입니다. 

```python
# 모델 테스트 및 예측
print("\n 테스트 정확도 : {:.2f}%".format(model.evaluate(X_test,y_test)[1]*100))
predict = model.predict(X_test)
```

```python
# 예측 결과 출력
predict_labels = np.argmax(predict, axis=1)
original_labels = np.argmax(y_test, axis=1)

for i in range(100):
  print("가사 : ", test_data['lyrics'].iloc[i], "/\t 원래 라벨 : ", original_labels[i], "/\t예측한 라벨 : ", predict_labels[i])
```

### 7. 결과 집계 및 해석

모델에 의해 분류된 음악 가사의 감성 라벨(predict_labels)을 집계하고, 각 음악 트랙에 대한 전반적인 감성 경향을 평가하는 것입니다.

```python
def divide_list(l, n): 
    # 리스트 l의 길이가 n이면 계속 반복
    for i in range(0, len(l)-3, n): 
        yield l[i:i + n] 
# 한 리스트에 몇개씩 담을지 결정
n = 36
result = list(divide_list(predict_labels, n))

def make_val ():
  b=[]
  for val in result:
    a=np.sum(val)
    if a > 40:
      b.append('긍정 노래입니다.')
    elif a < 40 :
      b.append('부정 노래입니다.')
    else:
      b.append('중립 노래입니다.')
  for i in range(len(b)):
    print("제목 : ", title_name['titles'].iloc[i], "/\t예측한 결과 : ", b[i])
```

```python
make_val()
```

### 8. 학습 과정 시각화

훈련 및 검증 데이터에 대한 손실(loss)과 정확도(accuracy)를 에포크별로 시각화합니다.

```python
# 학습 곡선 시각화
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], 'b-', label='Training Loss')
plt.plot(history.history['val_loss'], 'r--', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], 'g-', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.legend()

plt.show()
```

### 9. 새로운 가사에 대한 감정 예측

사용자로부터 입력받은 가사에 대해 감정을 예측하는 함수를 정의합니다. 이후, 가사를 입력하여 정확도와 긍정/중립/부정을 분석합니다.

```python
def predict_sentiment(lyric):

    okt = Okt()

    tokens = okt.morphs(lyric, stem=True)
    tokens = [token for token in tokens if token not in stopwords]

    sequence = tokenizer.texts_to_sequences([tokens])
    padded_sequence = pad_sequences(sequence, maxlen=max_len)

    prediction = model.predict(padded_sequence)
    label = "긍정" if np.argmax(prediction) == 2 else ("부정" if np.argmax(prediction) == 0 else "중립")
    probability = np.max(prediction) * 100

    return f"{probability:.2f}% 확률로 {label} 노래입니다."
```

```python
predict_sentiment("참 서툴고 어설픈 나지만")
```

```python
predict_sentiment("서로를 잃기 싫은 모습까지도  Oh oh oh")
```

```python
predict_sentiment("사랑해 너를 사랑해 죽을 만큼 힘이 들지만")
```
